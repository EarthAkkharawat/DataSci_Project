```python
import requests
import urllib.request
from PIL import Image
import matplotlib.pyplot as plt

classes = ['canal', 'electric', 'flooding', 'light', 'road', 'sanitary', 'sewer', 'sidewalk', 'stray', 'traffic']
headers = {
    'Content-Type': 'application/json'  # Replace with the correct content type
}
url_list = ["https://storage.googleapis.com/traffy_public_bucket/attachment/2023-05/9ff29d91d3c7d1494947e9ce624dfb67efe3f0b2.jpg",
            "https://storage.googleapis.com/traffy_public_bucket/attachment/2023-05/34475950ad005c4d40f8c19cf303e8172ea49e3e.jpg",
            "https://upload.wikimedia.org/wikipedia/commons/thumb/4/48/059_-_Marvel_Vs_DC_-_psantaana.jpg/1200px-059_-_Marvel_Vs_DC_-_psantaana.jpg"]
for url in url_list:
    with urllib.request.urlopen(url) as f:
        image = Image.open(f)
        
    payload = {
                "url":url
    }
                    
    response = requests.post("https://tofu-api-nj2eo5v2pq-as.a.run.app", json=payload, headers=headers)
    plt.imshow(image)
    plt.show()
    print(response.json()['prediction'])
    pred = ""
    for i in range(len(response.json()['prediction'])):
        if response.json()['prediction'][i] == 1: 
            pred+=classes[i]+", "
    print(pred[:-2])
    print("\n_______________________________________")
```
